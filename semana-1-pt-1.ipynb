```python
# Plot vecotrs, please keep the parameters in the same length
# @param: Vectors = [{"vector": vector variable, "name": name of vector, "color": color of the vector on diagram}]
    
from typing import List

def plotVec(vectors: List[dict]):
    ax = plt.axes()
    
    # For loop to draw the vectors
    for vec in vectors:
        ax.arrow(0, 0, *vec["vector"], head_width = 0.05,color = vec["color"], head_length = 0.1)
        plt.text(*(vec["vector"] + 0.1), vec["name"])
    
    plt.ylim(-2,2)
    plt.xlim(-2,2)

    plt.grid(True, which='both', linestyle='dotted', linewidth=0.5)
```

```python
# Create two sample tensors

u = torch.tensor([1, 0])
v = torch.tensor([0, 1])

# Add u and v

w = u + v
print("The result tensor: ", w)

# Plot u, v, w

plotVec([
    {"vector": u.numpy(), "name": 'u', "color": 'r'},
    {"vector": v.numpy(), "name": 'v', "color": 'b'},
    {"vector": w.numpy(), "name": 'w', "color": 'g'}
])
```

## O que é Média?

A média é uma medida estatística amplamente usada para representar um valor central em um conjunto de dados. É o resultado da soma de todos os valores dividido pelo número total de valores na amostra.

## Como Calcular a Média?

Para calcular a média de um conjunto de valores, siga estes passos:

1. **`Soma dos Valores`:** Some todos os valores presentes na amostra.

2. **`Número de Valores`:** Conte quantos valores existem no conjunto.

3. **`Cálculo da Média`:** Divida a soma dos valores pelo número total de valores.

## Interpretação da Média

A média representa um valor típico ou central em um conjunto de dados. Ela nos dá uma ideia geral de onde os valores estão localizados. Se alguns valores são maiores que a média, outros serão menores, o que equilibra a distribuição.

Por exemplo, considere o conjunto de valores: 5, 10, 15, 20 e 25. A soma dos valores é 75 e há 5 valores no conjunto. A média seria 75 dividido por 5, resultando em 15. Nesse caso, 15 representa o valor médio da amostra.

## Uso e Importância da Média

A média é uma medida de tendência central amplamente utilizada em diversas áreas, desde estatísticas básicas até análises complexas. É valiosa para resumir grandes conjuntos de dados em um único valor representativo.

Quando se lida com dados numéricos, a média é frequentemente o ponto de partida para entender as características do conjunto. Ela pode ser usada para comparar diferentes grupos, identificar tendências e padrões, ou mesmo para fazer previsões em algumas situações.

No entanto, é essencial ter em mente que a média pode ser afetada por valores extremos (outliers) que distorcem a representatividade. Nesses casos, outras medidas de tendência central, como a mediana, podem ser mais apropriadas.

Em resumo, a média é uma ferramenta valiosa para a análise de dados e é amplamente utilizada em vários campos, desde a vida cotidiana até pesquisas científicas e análises financeiras.

$$
\text{Média} = \overline{x} = \frac{\sum_{i=1}^{n} x_i}{n} = \frac{x_1 + ... + x_n}{n}
$$

## O que é Desvio Padrão?

O desvio padrão é uma medida estatística importante que nos ajuda a entender a dispersão ou variabilidade dos dados em uma amostra. Em outras palavras, o desvio padrão nos diz o quanto os valores da amostra se desviam da média.

## Como Calcular o Desvio Padrão?

O cálculo do desvio padrão envolve três passos principais:

1. **`Calcule a Média`:** Primeiro, somamos todos os valores da amostra e dividimos pelo número total de valores para obter a média.

2. **`Calcule a Diferença entre Cada Valor e a Média`:** Em seguida, para cada valor na amostra, subtraímos a média do valor correspondente. Essa diferença é chamada de "desvio" individual.

3. **`Calcule a Média dos Quadrados dos Desvios`:** Elevamos cada desvio ao quadrado e, em seguida, somamos todos os quadrados dos desvios. Dividimos esse resultado pelo número total de valores na amostra.

Finalmente, para obter o desvio padrão, tiramos a raiz quadrada da média dos quadrados dos desvios. Essa raiz quadrada é o desvio padrão, representado pelo símbolo grego sigma (σ).

## Interpretação do Desvio Padrão

Quanto maior o desvio padrão, mais dispersos os valores da amostra estão em relação à média. Por outro lado, um desvio padrão menor indica que os valores da amostra estão mais próximos da média, resultando em menor variabilidade.

O desvio padrão é uma ferramenta essencial em estatística, pois nos permite avaliar a incerteza nos dados e entender a consistência ou variabilidade de um conjunto de valores. Ele é amplamente utilizado em diversas áreas, como economia, ciências sociais, ciências naturais, entre outras.

Lembre-se de que o desvio padrão é uma medida válida apenas para amostras. Se estivermos trabalhando com uma população inteira, usamos uma fórmula ligeiramente diferente para calcular o desvio padrão populacional.

O cálculo do desvio padrão é um conceito central para qualquer análise estatística e nos proporciona uma visão valiosa sobre a distribuição dos dados em uma amostra.

$$
\text{Desvio Padrão} = \sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \overline{x})^2}
$$
